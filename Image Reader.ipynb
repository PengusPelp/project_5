```python
def rgb_array(np_array):
    arrays = []
    for i in np_array:
        for a in i:
            a = list(a[:3])
            arrays.append(a)
    return pd.DataFrame(arrays, columns = ['r','g','b'])
```

```python
df = rgb_array(img[:3000])
```
